```python
data = pd.read_csv("nba-players-data.csv")

# Drop rows with zero values
data = data[~(data == 0).any(axis=1)]

data_matrix = data.values 

# alternatives
m = data_matrix.shape[0]
# criterias
n = data_matrix.shape[1]

### get compare matrix from input file (only with low triangle)
compare_matrix = get_compare_matrix("weight.txt")
normalized_matrix = calculate_normalize_matrix(compare_matrix)

# sum of weight should be 1. weight for each criterion
weight_matrix = calculate_weight_matrix(normalized_matrix)
# print("weight for each criterion")
# print(weight_matrix)

# weighted sum
avg_weight_matrix = calculate_average_weight_matrix(compare_matrix, weight_matrix)
lamda = calculate_lamda(weight_matrix, avg_weight_matrix)
CI = calculate_CI(lamda, n)
RI = calculate_RI(n)
CR = calculate_CR(CI, RI)

# create and noralize alternative matrics
alternatives_matrices = get_alternative_matrics(n,m, data_matrix)
alternatives_norm_matrices = get_alternative_norm_matrics(alternatives_matrices)

# weight for each alternative with respect to each criteria
avg_weight_alternatives = get_avg_weight_alternatives(alternatives_norm_matrices)

# final score for each alternative
scores = np.dot(avg_weight_alternatives, weight_matrix)
# print("scores")
# print(scores)
```

```python
# display results
data['score'] = scores 
data.sort_values("score", ascending=False, inplace=True)
ranking = [(i+1) for i in range(len(scores))]
data['ranking'] = ranking
data.head(10)
```
